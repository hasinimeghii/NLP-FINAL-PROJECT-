# Topic Modeling
Using LDA to extract topics from customer reviews.

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

df = pd.read_csv('../data/dataset_enriched.csv')
texts = df['cleaned_text'].dropna().astype(str).tolist()[:5000] # Subset for speed

# Vectorize
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words=['le','la','les','et','de','des','un','une','pour','en','que','qui','dans','sur','pas','plus'])
tf = vectorizer.fit_transform(texts)
```

## 1. LDA Training

```
n_topics = 5
lda = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda.fit(tf)

feature_names = vectorizer.get_feature_names_out()
topics = {}

print("Extracted Topics:")
for topic_idx, topic in enumerate(lda.components_):
    top_features_ind = topic.argsort()[:-10 - 1:-1]
    top_features = [feature_names[i] for i in top_features_ind]
    topics[f"Topic {topic_idx+1}"] = top_features
    print(f"Topic {topic_idx+1}: {', '.join(top_features)}")
```

## 2. Topic Output & Naming
Based on words, we assign manual names.

```
# Manually derived labels based on consistent patterns in runs:
named_topics = {
    'Topic 1': 'Contract/Subscription Duration',
    'Topic 2': 'Claims and Compensation',
    'Topic 3': 'Capital & Policy Details',
    'Topic 4': 'Costs & Reimbursements',
    'Topic 5': 'Customer Service Issues'
}

for k, v in named_topics.items():
    print(f"{k} refers to: {v}")
```
