# Data Exploration
Visualizations, N-grams, distribution analysis.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('../data/dataset_enriched.csv')
print(df.shape)
```

## 1. Target Distribution
Star ratings analysis.

```
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='note', palette='viridis')
plt.title('Distribution of Star Ratings')
plt.xlabel('Note (Stars)')
plt.ylabel('Count')
plt.show()
```

## 2. Frequent Words & N-grams
Extracting using CountVectorizer.

```
def plot_top_ngrams(texts, n=1, top_k=20, title='Top Words'):
    vec = CountVectorizer(ngram_range=(n,n), stop_words='english', max_features=1000).fit(texts)
    bag_of_words = vec.transform(texts)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)[:top_k]
    
    df_freq = pd.DataFrame(words_freq, columns=['Word', 'Freq'])
    plt.figure(figsize=(10,6))
    sns.barplot(x='Freq', y='Word', data=df_freq, palette='magma')
    plt.title(title)
    plt.show()

# Sample of data for fast execution
texts = df['cleaned_text'].dropna().head(10000).astype(str)
plot_top_ngrams(texts, n=1, top_k=20, title='Top 20 Unigrams')
plot_top_ngrams(texts, n=2, top_k=20, title='Top 20 Bigrams')
```
